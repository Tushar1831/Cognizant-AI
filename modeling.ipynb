Need to fix timestamp column in all datasets

```python
# Creating formatting function
def convert_to_dateime(df: pd.DataFrame = None, column: str = None):
    """Converts timestamp column to datetime datatype in format '%Y-%m-%d %H:%M:%S'

    Args:
        df (pd.DataFrame, optional): Data in the form of a pandas DataFrame. Defaults to None.
        column (str, optional): Timestamp column name in string format. Defaults to None.
    """
    
    dummy = df.copy()
    dummy[column] = pd.to_datetime(dummy[column], format = '%Y-%m-%d %H:%M:%S')
    return dummy
```

<h1> Merge Data</h1>

<h3>Convert timestamp to hourly</h3>

```python
from datetime import datetime

def convert_timestamp_to_hourly(df: pd.DataFrame = None, column: str = None):
    """Converts timestamp column to hourly

    Args:
        df (pd.DataFrame, optional): data in the form of pandas DataFrame. Defaults to None.
        column (str, optional): column name in string format. Defaults to None.
    """
    
    dummy = df.copy()
    ts = dummy[column].tolist()
    ts = [i.strftime('%Y-%m-%d %H:00:00') for i in ts]
    ts = [datetime.strptime(i, '%Y-%m-%d %H:00:00') for i in ts]
    dummy[column] = ts
    
    return dummy
```

```python
sales_df = convert_timestamp_to_hourly(sales_df, 'timestamp')
sales_df.head()
```

```python
stock_df = convert_timestamp_to_hourly(stock_df, 'timestamp')
stock_df.head()
```

```python
temp_df = convert_timestamp_to_hourly(temp_df, 'timestamp')
temp_df.head()
```

```python
sales_agg = sales_df.groupby(['timestamp', 'product_id']).agg({'quantity': 'sum'}).reset_index()
sales_agg.head()
```

```python
stock_agg = stock_df.groupby(['timestamp', 'product_id']).agg({'estimated_stock_pct': 'mean'}).reset_index()
stock_agg.head()
```

```python
temp_agg = temp_df.groupby(['timestamp']).agg({'temperature': 'mean'}).reset_index()
temp_agg.head()
```

<h3>Merging Tables</h3>

```python
merged_df = stock_agg.merge(sales_agg, on=['timestamp', 'product_id'], how='left')
merged_df.head()
```

```python
merged_df = merged_df.merge(temp_agg, on='timestamp', how='left')
merged_df.head()
```

```python
merged_df.info()
```

There are null values in the quantity column. This indicates that there were no sales of the product in that hour hence replace the null values with 0.

```python
merged_df['quantity'] = merged_df['quantity'].fillna(0)
merged_df.info()
```

Adding additional features including product category and price.

```python
product_categories = sales_df[['product_id', 'category']].drop_duplicates()

product_price = sales_df[['product_id', 'unit_price']].drop_duplicates()
```

```python
merged_df = merged_df.merge(product_price, on='product_id', how = 'left')
merged_df.head()
```

```python
merged_df = merged_df.merge(product_categories, on='product_id', how = 'left')
merged_df.head()
```

```python
merged_df = merged_df.drop('category_y', axis = 'columns')
merged_df.head()
```

```python
merged_df.columns = ['timestamp', 'product_id', 'estimated_stock_pct', 'quantity', 'temperature', 'unit_price', 'category']
merged_df.info()
```
