Need to fix timestamp column in all datasets

```python
# Creating formatting function
def convert_to_dateime(df: pd.DataFrame = None, column: str = None):
    """Converts timestamp column to datetime datatype in format '%Y-%m-%d %H:%M:%S'

    Args:
        df (pd.DataFrame, optional): Data in the form of a pandas DataFrame. Defaults to None.
        column (str, optional): Timestamp column name in string format. Defaults to None.
    """
    
    dummy = df.copy()
    dummy[column] = pd.to_datetime(dummy[column], format = '%Y-%m-%d %H:%M:%S')
    return dummy
```

```python
sales_df = convert_to_dateime(sales_df, 'timestamp')
sales_df.info()
```

<h3>Convert timestamp to hourly</h3>

```python
from datetime import datetime

def convert_timestamp_to_hourly(df: pd.DataFrame = None, column: str = None):
    """Converts timestamp column to hourly

    Args:
        df (pd.DataFrame, optional): data in the form of pandas DataFrame. Defaults to None.
        column (str, optional): column name in string format. Defaults to None.
    """
    
    dummy = df.copy()
    ts = dummy[column].tolist()
    ts = [i.strftime('%Y-%m-%d %H:00:00') for i in ts]
    ts = [datetime.strptime(i, '%Y-%m-%d %H:00:00') for i in ts]
    dummy[column] = ts
    
    return dummy
```

```python
sales_df = convert_timestamp_to_hourly(sales_df, 'timestamp')
sales_df.head()
```

```python
stock_df = convert_timestamp_to_hourly(stock_df, 'timestamp')
stock_df.head()
```

```python
temp_df = convert_timestamp_to_hourly(temp_df, 'timestamp')
temp_df.head()
```
